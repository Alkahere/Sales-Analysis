```python
import pandas as pd
import os
from os import listdir
```

```python
df = pd.read_csv("./SalesAnalysis/Sales_Data/Sales_April_2019.csv")

files= [file for file in os.listdir('./SalesAnalysis/Sales_Data')]

all_months_data = pd.DataFrame()

for file in files:
    df = pd.read_csv("./SalesAnalysis/Sales_Data/"+file)
    all_months_data = pd.concat([all_months_data , df])
    
    all_months_data.to_csv("all_data_csv", index =False)
    all_months_data.head()
```

```python
all_months_data['Months']=3
all_months_data.head()
```

```python
all_months_data['Months']=all_months_data['Order Date'].str[0:2]
all_months_data.head()
```

```python
nan_def = all_months_data[all_months_data.isna().any(axis=1)] 
nan_def.head()
```

```python
nan_def = all_months_data[all_months_data.isna().any(axis=1)] 
nan_def.head()

all_months_data=all_months_data.dropna(how='all')
all_months_data.head()
```

```python
all_months_data=all_months_data[all_months_data['Order Date'].str[0:2] !='Or']
all_months_data.head()                                
                                
```

```python
all_months_data['Quantity Ordered']= pd.to_numeric(all_months_data['Quantity Ordered']) #make int
all_months_data['Price Each']= pd.to_numeric(all_months_data['Price Each'])#make float

all_months_data.head()
```

```python
all_months_data.loc[:,'Month'] = all_months_data['Order Date'].str[0:2]
all_months_data.loc[:, 'Month'] = all_months_data['Month'].astype('int32')

all_months_data.head()
```

```python
all_months_data.loc[:, 'Sales'] = all_months_data['Quantity Ordered'] * all_months_data['Price Each']

all_months_data.head()
```
